# Importando bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
```

# Importando Base de dados e transformando em um DataFrame

```python
df = pd.read_excel('arquivo_bom.xlsx')
df
```

# Eliminando linhas com o vencimento zerado

```python
df.drop(df[df['VENCIMENTOS'] == 0].index, inplace=True)
# df=df.drop(df[df['VENCIMENTOS'] == 0].index)
```

# Fazendo o gráfico

```python

# Suponha que você tenha um DataFrame chamado 'df' com duas colunas que deseja plotar
df1 = df.iloc[:, [7,8]]

# Plotando as barras para 'Valor1'
plt.bar(df1['NASCIMENTO'], df1['VENCIMENTOS'], width=1, label='Salário', align='center')


# Adicionando título e rótulos
plt.title('Gráfico de Barras')
plt.xlabel('Ano de Nascimento')
plt.ylabel('Valores')

# Adicionando legenda
plt.legend()

# Exibindo o gráfico
plt.show()
```

# Moda dos salários

```python
moda_dos_salarios=st.mode(df1['VENCIMENTOS'])
moda_dos_salarios
```

# Mediana dos salários

```python
mediana_dos_salarios=st.median(df1['VENCIMENTOS'])
mediana_dos_salarios
```

# Média dos salários

```python
medias_dos_salarios=np.mean(df1['VENCIMENTOS'])
medias_dos_salarios
```

# Variância dos salários

```python
# identificando que a dispersão dos valores desta base é imensa.
desvio_padrao = st.stdev(df1['VENCIMENTOS'])
desvio_padrao
```
